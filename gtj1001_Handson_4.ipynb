## 6.1 Prime Number Checker
Create a program that checks whether a number is a prime number and displays its factors if it is not a prime number.

### Console:
```powershell
Prime Number Checker

Please enter an integer between 1 and 5000: 5
5 is a prime number.

Try again? (y/n): y

Please enter an integer between 1 and 5000: 6
6 is NOT a prime number.
It has 4 factors: 1 2 3 6

Try again? (y/n): y

Please enter an integer between 1 and 5000: 200
200 is NOT a prime number.
It has 12 factors: 1 2 3 4 5 8 10 20 25 40 50 100 200

Try again? (y/n): n

Bye!
```

### Specifications:
- A prime number is divisible by two factors (1 and itself). For example, 7 is a prime number because it is only divisible by 1 and 7.
- If the user enters an integer that's not between 1 and 5000, the program should display an error message.
- If the number is a prime number, the program should display a message.
If the number is not a prime number, the program should display a message. Then, it should display the number of factors for the number and a list of those factors.
- Store the factors for each number in a list.
- Use functions to organize the code for this program.


````python
def factorCalc(x):
    count=0
    factorList=[]
    for i in range(1,x+1):
        if x%i==0:
            count+=1
            factorList.append(i)
    return count,factorList
def entry():
    while True:
        x=int(input("Please enter an integer between 1 and 5000: "))
        if x>=1 and x<=5000:
            break
        else:
            print("Invalid entry, please try again.")
    return x
def main():
    print("Prime Number Checker")
    print()
    choice="y"
    while choice.lower()=="y":
        x=entry()
        y,factors=factorCalc(x)
        if y==2:
            print(str(x)+" is a prime number.")
        elif y==1:
            print(str(x)+" is neither prime nor composite.")
        else:
            print(str(x)+" is NOT a prime number.")
            print("It has "+str(y)+" factors:",end=' ')
            for i in factors:
                print(str(i),end=' ')
            print()
        print()
        choice=input("Try again? (y/n): ")
    print("\nBye!")
if __name__=="__main__":
    main()
````

## 6.2 - Contact Manager
Create a program that a user can use to manage the primary email address and phone number for a contact.

### Console
```powershell
Contact Manager

COMMAND MENU
list - Display all contacts
view - View a contact
add - Add a contact
del - Delete a contact
exit - Exit program

Command: list
1. Guido van Rossum
2. Eric Idle

Command: view
Number: 2
Name: Eric Idle
Email: eric@ericidle.com
Phone: +44 20 7946 0958

Command: add
Name: Mike Murach
Email: mike@murach.com
Phone: 559-123-4567
Mike Murach was added.

Command: del
Number: 1
Guido van Rossum was deleted.

Command: list
1. Eric Idle
2. Mike Murach

Command: exit
Bye!
```

### Specifications:
- Use a list of lists to store the data for the contacts. Provide starting data for two or more contacts.
- For the `view` and `del` commands, display an error message if the user enters and invalid contact number.
- When you exit the program, all changes that you made to the contact list are lost.




````python
def list(contact_list):
    if len(contact_list)==0:
        print("The list is empty\n")
    else:
        i=1
        for item in contact_list:
            print(str(i)+" "+item[0])
            i+=1
        print()
def view(contact_list):
    if contact_list==[]:
        print("The list is empty\n")
    else:
        while True:
            entry=int(input("Number: "))
            if entry>len(contact_list):
                print("Invalid index number.")
                continue
            else:
                print("Name: ",contact_list[entry-1][0])
                print("Email: ",contact_list[entry-1][1])
                print("Phone: ",contact_list[entry-1][2])
                print()
                break
def add(contact_list):
    new_contact=[]
    name=input("Name: ")
    email=input("Email: ")
    phone=input("Phone: ")
    new_contact.append(name)
    new_contact.append(email)
    new_contact.append(phone)
    contact_list.append(new_contact)
    print(new_contact[0]," was added.")
    print()
def delete(contact_list):
    entry=int(input("Number to delete: "))
    if entry<1 or entry>len(contact_list):
        print("Invalid entry.")
        print()
    else:
        deleted_contact=contact_list.pop(entry-1)
        print(deleted_contact[0]+" was deleted.")
        print()
def menu():
    print("=================")
    print("COMMAND MENU:")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("delete - Delete a contact")
    print("exit - Exit program")
    print("=================")
    print()
def main():
    contact_list=[["Guido van Rossum","guido@rossum.van","123 456 7890"],
                  ["Eric Idle","eric@ericidle.com","+44 20 7946 0958"]]
    menu()
    while True:
        command=input("Command: ")
        if command=="list":
            list(contact_list)
        if command=="view":
            view(contact_list)
        if command=="add":
            add(contact_list)
        if command=="delete":
            delete(contact_list)
        if command=="exit":
            break
    print("Bye!")
if __name__=="__main__":
    main()
````
