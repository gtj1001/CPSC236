Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:

Germany	        4       1954, 1974, 1990, 2014

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,Jos√© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
with open("world_cup_champions.txt") as file:
    champions=file.read()
champlist=champions.split("\n")
#champlist is a non nested list where every item is a comma separated string

formated=[]
i=0
while i<len(champlist):
    formated.append(champlist[i].split(","))
    i+=1
formated.pop(0)
#formated is a list of lists without the key line

dictionary={}
i=0
while i<len(formated):
    data_to_add={}
    if formated[i][1] in dictionary:
        win_count=dictionary[formated[i][1]]["win_count"]
        win_count+=1
        data_to_add["win_count"]=win_count
        years=dictionary[formated[i][1]]["years"]
        years.append(formated[i][0])
        data_to_add["years"]=years
    else:
        data_to_add["win_count"]=1
        years=[]
        years.append(formated[i][0])
        data_to_add["years"]=years
    data_to_add["coach"]=formated[i][2]
    data_to_add["captain"]=formated[i][3]
    dictionary[formated[i][1]]=data_to_add
    i+=1
country_list=list(dictionary.keys())
country_list.sort()
print("Country\t\tWins\tYears")
print("=======\t\t====\t=====")
for item in country_list:
    item_wins=dictionary[item]["win_count"]
    item_years=dictionary[item]["years"]
    if len(item)>8:
        print(item,end="\t")
    else:
        print(item,end="\t\t")
    print(item_wins,end="\t")
    for i in item_years:
        print(i,end=" ")
    print()

#I have no idea how I managed to make this work first try...
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

COMMAND MENU

edit	- Edit sales for specified month

Sales amount for Jan is 14,317.00.

Command: edit

Sales Amount: 15293
Sales amount for Jan is 15,293.00.

Command: totals
Yearly total: 67,855.00
Monthly average: 0

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
def menu():
    print("COMMAND LIST")
    print("view\t-View Sales for specified month")
    print("edit\t-Edit sales for specified month")
    print("totals\t-View sales summary for year")
    print("exit\t-Exit program")

def initialize_sales():
    with open("monthly_sales.txt") as file:
        sales=file.read()
    sales=sales.split("\n")
    formated_sales=[]
    for item in sales:
        if len(item)==0:
            break
        else:
            item=item.split("\t")
            formated_sales.append(item)
    dictionary=dict(formated_sales)
    return dictionary

def view_sales(dictionary):
    print()
    month=input("Three letter month: ")
    if month.title() in dictionary:
        print("Sales amount for "+month.title()+" is "+dictionary.get(month.title()))
    else:
        print("Invalid month entry")
    print()
    
def edit_sales(dictionary):
    print()
    month=input("Three letter month: ")
    month=month.title()
    sales=input("Sales amount: ")
    dictionary[month]=sales
    with open("monthly_sales.txt","w") as file:
        for item in dictionary:
            file.write(item+"\t"+dictionary[item]+"\n")
    print("New sales amount for "+month+" is "+sales+".")
    

def sales_summary(dictionary):
    total=0
    for item in dictionary:
        total+=int(dictionary[item])
    print("\nYearly total: "+str(round(total,2)))
    average=round(total/12,2)
    print("Monthly average: "+str(average))
    
def main():
    dictionary=initialize_sales()
    print("Monthly Sales Program\n")
    menu()
    while True:
        command=input("Command: ")
        if command=="view":
            view_sales(dictionary)
        elif command=="edit":
            edit_sales(dictionary)
        elif command=="totals":
            sales_summary(dictionary)
        elif command=="exit":
            print("Bye!")
            break
        else:
            print("Invalid command.")

if __name__=="__main__":
    main()
````
